```python
import os
import shutil
```

```python
src_pointclouds = r"D:\GitHubProjects\Terranum_repo\pc_movement_tracking\data\test_14_neigh_on_target_2\results\pointclouds"
x_min = 312
x_max = 313
y_min = 312
y_max = 313
src_res = os.path.join(src_pointclouds, f'target_region_x_btw_{x_min}_{x_max}_y_btw_{y_min}_{y_max}')
os.makedirs(src_res, exist_ok=True)

list_files = [x for x in os.listdir(src_pointclouds) if x.endswith('.ply')]
for file in list_files:
    x = float(file.split('x=')[1].split('_')[0])
    y = float(file.split('y=')[1].split('_')[0])
    if x > x_min and x < x_max and y > y_min and y < y_max:
        lvl = file.split("lvl=")[1].split('_')[0]
        print(f"lvl={lvl}", x, '-', y)
        shutil.copyfile(
            os.path.join(src_pointclouds, file),
            os.path.join(src_res, file),
        )
```
